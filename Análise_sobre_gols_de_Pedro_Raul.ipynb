### 1 - Quais as datas do primeiro e do último gol de Pedro Raul?

### 2 - Quais os principais tipos de gols de Pedro Raul?

```python
# Agrupamento da quantidade de gols por tipo
gols = df.groupby(['Tipo de gol']).size().reset_index(name = 'quantidade')
gols
# Ordenamento dos valores por 'quantidade' em decrescente
gols = gols.sort_values(by = 'quantidade', ascending = False)

# Para plotar os gráficos de barras
plt.figure(figsize=(12, 8))
plt.bar(gols['Tipo de gol'], gols['quantidade'], color = 'black')
plt.xlabel('Tipo de gol')
plt.ylabel('Quantidade')
plt.title('Quantidade de gols por tipo')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### 3 - Quais foram os times que Pedro Raul mais marcou gols?

```python
times = df['Time'].value_counts().reset_index()
times.columns = ['Time', 'quantidade']

# Para filtrar e selecionar o 3 times que ele mais atuou, marcando gol por eles
top_3_times = times.nlargest(3, 'quantidade').sort_values(by='quantidade', ascending=False)

# Plotagem do gráfico
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(top_3_times['Time'], top_3_times['quantidade'])
ax.set_title('Times de sua carrreira')
ax.set_xlabel('Clube')
ax.set_ylabel('Gols')
plt.show()
```

### 4 -  Quais são os 5 principais assistentes de Pedro Raul?

```python
# Plotagem do gráfico de barras horizontais

# Para contar o número de assistencias para gols
assistencia = df['Assistência para gol'].value_counts().reset_index()
assistencia.columns = ['Assistência para gol', 'quantidade']

# Para selecionar os 5 principais assistentes
top_5_assistencia = assistencia.nlargest(5, 'quantidade').sort_values(by='quantidade', ascending=True)

fig, (ax2) = plt.subplots(1, figsize=(24, 6), dpi=72)

# Plotagem do gráfico dos principais assistentes
ax2.barh(top_5_assistencia['Assistência para gol'], top_5_assistencia['quantidade'], color='#4567')
ax2.set_xlabel('Assistente')
ax2.set_ylabel('Gols Sofridos')
ax2.text(0.45, 1.05,'5 maiores assistentes', fontsize=22, color='gray',fontweight='bold', transform = ax2.transAxes)
plt.show()
```

### 5 - Quais são os times que mais levaram gols de Pedro Raul?

```python
# Plotagem do gráfico de barras verticais

# Para contar o número de gols sofridos por cada clube
gols_sofridos = df['Adversário'].value_counts().reset_index()
gols_sofridos.columns = ['Adversário', 'quantidade']

# Para selecionar os 5 principais clubes que levaram gols
times_gols_sofridos = gols_sofridos.nlargest(5, 'quantidade').sort_values(by='quantidade', ascending=True)

# Plotagem do gráfico de clubes que mais levaram gols
fig, ax = plt.subplots(figsize=(16, 6))
ax.bar(times_gols_sofridos['Adversário'], times_gols_sofridos['quantidade'])
ax.set_title('Times de sua carrreira')
ax.set_xlabel('Clube')
ax.set_ylabel('Gols')
plt.show()
```

### 6 - Quais as faixas de tempo de todos os gols?

```python
# Para retirar o símbolo '+' e qualquer coisa que apareça após ele na coluna 'minuto_normalizado'
df['minuto_normalizado'] = df['Minuto'].str.replace(r'\+.*$', '', regex=True)

# Para retirar qualquer caractere não numérico da coluna 'minuto_normalizado'
df['minuto_normalizado'] = df['minuto_normalizado'].str.replace(r'\D', '', regex=True)

# Para alterar o campo 'minuto_normalizado' para o tipo inteiro
df['minuto_normalizado'] = df['minuto_normalizado'].astype(np.int64)

df.info()
```

```python
# Para determinar os rótulos para os ranges de tempo
def segmentar_faixa_tempo(valor):
    if valor <= 15:
        return '0-15'
    elif valor <= 30:
        return '15-30'
    elif valor <= 45:
        return '30-45'
    elif valor <= 60:
        return '45-60'
    elif valor <= 75:
        return '60-75'
    else:
        return '75-90'

# Para executar a função aos valores da coluna 'minuto_normalizado' para criar os intervalos de tempo
df['intervalo_tempo'] = df['minuto_normalizado'].apply(segmentar_faixa_tempo)

# Para apresentar as linhas em cada intervalo de tempo
contagem_por_intervalo = df['intervalo_tempo'].value_counts().sort_index()

print(contagem_por_intervalo)
```

```python
# Plotagem do gráfico de barras
plt.figure(figsize=(10, 6))
contagem_por_intervalo.plot(kind='bar', color='black')
plt.title('Contagem de Gols por Range de Tempo da partida')
plt.xlabel('Intervalo de Tempo')
plt.ylabel('Quantidade de gols')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.9)
plt.show()
```

# 9 - Quais os resultados finais dos jogos em que Yuri fez gol?

```python
# Função criada para extrair os gols do time mandante
def extrair_gols_mandante(Local, Resultado):
    if Local == 'F':
        return Resultado.split(':')[0]
    elif Local == 'C':
        return Resultado.split(':')[0]
    else:
        return None

# Função criada para extrair os gols do time visitante
def extrair_gols_visitante(Local, Resultado):
    if Local == 'F':
        return Resultado.split(':')[1]
    elif Local == 'C':
        return Resultado.split(':')[1]
    else:
        return None

# Aplicar as funções às colunas 'cidade' e 'resultado' para criar as novas colunas
df['gols_mandante'] = df.apply(lambda x: extrair_gols_mandante(x['Local'], x['Resultado']), axis=1)
df['gols_visitante'] = df.apply(lambda x: extrair_gols_visitante(x['Local'], x['Resultado']), axis=1)

# Função para determinar o resultado final do jogo
def determinar_resultado_final(Local, gols_mandante, gols_visitante):
    if Local == 'C':
        if gols_mandante > gols_visitante:
            return 'vitória'
        elif gols_mandante < gols_visitante:
            return 'derrota'
        else:
            return 'empate'
    elif Local == 'F':
        if gols_mandante > gols_visitante:
            return 'derrota'
        elif gols_mandante < gols_visitante:
            return 'vitória'
        else:
            return 'empate'
    else:
        return None

# Aplicar a função às colunas 'cidade', 'gols_mandante' e 'gols_visitante' para criar a nova coluna
df['resultado_final'] = df.apply(lambda x: determinar_resultado_final(x['Local'], x['gols_mandante'], x['gols_visitante']), axis=1)

# Contar o número de ocorrências de cada categoria em 'resultado_parcial'
contagem_resultado_final = df['resultado_final'].value_counts()

# Plotar o gráfico de barras
contagem_resultado_final.plot(kind='bar', color='yellow')

# Adicionar rótulos e título ao gráfico
plt.xlabel('Resultado Final')
plt.ylabel('Quantidade de gols')
plt.title('Quantidade de gols por Resultado Final')
plt.xticks(rotation=90)
# Exibir o gráfico
plt.show()
```
